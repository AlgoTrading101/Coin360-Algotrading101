```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

options = Options()
options.headless = True

# Change this to your dir and add it to PATH
DRIVER_PATH = 'C:/Users/igorr/Desktop/chromed/chromedriver.exe'

driver = webdriver.Chrome(options=options, executable_path=DRIVER_PATH)
driver.get("https://coin360.com")
```

# Obtain Coin Data

```python
data = driver.find_elements_by_xpath('//div[@class="Leaf"]')

parsed = []
for i in range(len(data)):
    parsed.append(data[i].text)
    print(data[i].text)
```

```python
parsed
```

```python
import pandas as pd

cols = ['symbol', 'price', 'pct_change']
values = []

for row in parsed:
    row = row.split('\n')
    values.append(row)
    
df = pd.DataFrame(values, columns=cols)
df.head()
```
